```python
# I have some trends data per year, I want to create a graph that shows the trend of the data
trends = {
    2024: 443306,
    2023: 2433257,
    2022: 2251027,
    2021: 2045263,
    2020: 1435839,
    2019: 1223821,
    2018: 929167,
    2017: 712956,
    2016: 542799,
    2015: 364961,
    2014: 230416,
    2013: 170789,
    2012: 126700,
    2011: 106283,
    2010: 86151,
    2009: 49516,
    2008: 25684,
    2007: 12478,
    2006: 7563,
    2005: 9067,
    2004: 1,
}
```

```python
from matplotlib import pyplot as plt

years = sorted(trends.keys())
values = [trends[year] for year in years]

cumulative_values = [sum(values[:i+1]) for i in range(len(values))]

plt.figure(figsize=(10, 6))
plt.plot(years, cumulative_values, marker='o')
plt.title('Cumulative number of packages on Maven Central')
plt.xlabel('Year')
plt.ylabel('Total number of packages')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x/1e6)}M' if x >= 1e6 else f'{int(x/1e3)}K' if x >= 1e3 else str(x)))
plt.tight_layout()

plt.savefig('cumulative_packages.pdf')


plt.show()


# save it as pdf
```
